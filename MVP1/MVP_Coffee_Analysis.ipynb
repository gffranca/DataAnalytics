### **1.1 Introdução**



Os cafés especiais representam um universo distinto no cenário brasileiro, contrastando diretamente com a familiaridade do café tradicional consumido pela maioria. Enquanto o café convencional utiliza uma mistura que inclui até 20% de defeitos e grãos Robusta para diminuir custos, os cafés especiais são exclusivamente produzidos com grãos Arábica selecionados, passando por um rigoroso processo de avaliação da SCAA, onde são submetidos a 10 critérios distintos e precisam atingir no mínimo 80 pontos em uma escala de 100 para serem considerados como tal.

A diferença entre os dois tipos de café torna-se evidente ao degustá-los. Enquanto o café tradicional tende a apresentar amargor acentuado e características indesejáveis devido à inclusão de grãos inferiores, os cafés especiais oferecem uma variedade de sabores e aromas distintos, com um notável equilíbrio, permitindo explorar nuances através de diferentes curvas de torra.

A Associação Brasileira da Indústria do Café (ABIC) classifica o café tradicional como a opção do dia a dia, acessível e certificado por eles, composto por grãos Arábica, Robusta e até 20% de defeitos, enquanto os cafés especiais são identificados como produtos de alta qualidade, exigindo um padrão mais elevado em todos os processos, desde o cultivo até o preparo. Assim, fica claro que o café vai além de uma simples bebida energética, exigindo conhecimento especializado e perpassando uma cadeia de produção meticulosa para proporcionar uma experiência sensorial e aromática única.

O trabalho de um provador de café é crucial, mas desafiador. Eles recebem amostras de café já torrado, moem e preparam pequenas xícaras para degustação. Avaliam o aroma seco, adicionam água quente para liberar sabores, formando uma crosta na superfície. Analisam o aroma dessa crosta e misturam para avaliar uniformemente.

Após higienizar, chegam à análise sensorial. Com uma técnica específica, absorvem um pouco do líquido para experimentar a temperatura certa e a explosão de sabores. O provador precisa identificar cada característica do café, fazendo isso repetidamente, especialmente em períodos de safra, podendo avaliar até 800 a 1000 lotes por dia em cerca de 5 xícaras cada.

Para evitar a overdose de cafeína, eles não bebem, mas sim cospem o café após a prova. É um trabalho exigente, que ainda continua com a avaliação de características do café como o aroma, sabor, acidez e corpo.

````python
# Formatação dos atributos de notas sensoriais
FLAVORS_HIERARCHY_LEVELS = 3 # Quantidade de níveis de flavors
FLAVORS_PREFIX_STRINGS = ['Group-', 'SubGroup-', 'Attribute-'] # Prefixo de atributo para cada nível
````

#### **2.4.2 Tratamento**

Apesar das funções de Tratamento parecerem com as de Manipulação, semanticamente essas funções apenas fazem um tratamento no formato dos dados, preparando-os de maneira específica para cada situação necessária.

A função ***convert_list_itens_to_lower*** recebe uma lista de strings e retorna uma lista de strings minúsculas:

````python
def convert_list_itens_to_lower(src: list[str]) -> list[str]:
    return [x.lower() for x in src]
````

A função ***treat_single_quotes*** recebe uma string e remove apostrofes dela:

````python
def treat_single_quotes(text: str) -> str:
    return text.replace('’', '').replace('\'', '')
````

A função ***format_flavors_by_hierarchy*** recebe uma lista de strings de notas sensoriais com o prefixo repectível do nível, e retorna-as com o seu prefixo:

````python
def format_flavors_by_hierarchy(flavors: list[str], hierarchy_level: int) -> list[str]:
    if hierarchy_level < 0 or hierarchy_level >= len(FLAVORS_PREFIX_STRINGS):
        return flavors

    prefix = FLAVORS_PREFIX_STRINGS[hierarchy_level]
    return [prefix + flavor for flavor in flavors]
````

#### **2.4.3 Manipulação**

De maneira geral, as funções de Manipulação relacionam, alteram e gerenciam dados. A maioria das funções e as mais utilizadas neste trabalho são deste tipo.

A função ***count_nonzero_lines_by_columns*** recebe um DataFrame, e para cada coluna nele, conta a quantidade de valores não zero e retorna um dicionário com essas informações:

````python
def count_nonzero_lines_by_columns(df: pd.DataFrame) -> dict[str, int]:
    contagem = {}
    for coluna in df.columns:
        contagem[coluna] = np.count_nonzero(df[coluna])
    return contagem
````

A função ***get_countries_from_text*** recebe a lista de países e regiões, e para cada série do DataFrame retorna a lista de regiões que se encontram no valor da coluna *beans_origin*:

````python
def get_countries_from_text(row: pd.Series, countries: list[str]) -> list[str]:
    regions = []

    # Remove apostrofes para facilitar identificação de países
    text = treat_single_quotes(str(row['beans_origin']).lower())

    # Para país possível, verifica se esta presente na região de origem e adicione a lista de origens
    for token in convert_list_itens_to_lower(countries):
        token = treat_single_quotes(token)
        if token in text:
            regions.append(token)

    regions = list(set(regions))
    return regions
````

A função ***get_origin_country_or_countries*** recebe a lista de países e regiões, e para cada série do DataFrame retorna uma tupla com o país de origem (se for de origem única) e a quantidade de países de origem:

````python
def get_origin_country_or_countries(row: pd.Series, countries: list[str]) -> tuple[str, int]:
    origins = get_countries_from_text(row, countries)
    return origins[0] if len(origins) == 1 else '', len(origins)
````

A função ***check_blend***, para cada série do DataFrame, verifica se corresponde a um blend e retorna um valor binário:

````python
def check_blend(row: pd.Series) -> int:
    return 1 if (any('blend' in str(value).lower() for value in row) or row['qtd_origin_countries'] == 0) else 0
````

A função ***check_espresso***, para cada série do DataFrame, verifica se corresponde a um espresso e retorna um valor binário:

````python
def check_espresso(row: pd.Series) -> int:
    return 1 if (any('espresso' in str(value).lower() for value in row) or row['with_milk'] >= 0) else 0
````

A função ***get_mean_sliding_window*** recebe um DataFrame e de acordo com o tamanho da janela de média móvel, retorna um DataFrame com o valor médio do período informado para cada ano:

````python
# Formato do DataFrame deve ser:
# Colunas: Região, ano1, ano2, ano3, ...
def get_mean_sliding_window(climate_dataset: pd.DataFrame, window_size: int) -> pd.DataFrame:
    # DataFrame vazio para armazenar as médias dos anos anteriores para cada coluna
    avg_data = pd.DataFrame(columns=climate_dataset.columns)

    # Define os valores da coluna 'Regiao' do novo DataFrame usando a mesma coluna do DataFrame original
    avg_data['region'] = climate_dataset['region']

    # Para cada coluna de ano, calcule a média dos anos anteriores e adicione ao novo DataFrame
    for col in avg_data.columns[1:]:
        window_start_col = max(1, climate_dataset.columns.get_loc(col) - window_size)
        window_end_col = climate_dataset.columns.get_loc(col)

        # Calcula a média dos anos anteriores para cada coluna
        avg_data[col] = climate_dataset.iloc[:, window_start_col:window_end_col].mean(axis=1)

    return avg_data
````

A função ***get_climate_value_of_region***, para cada série do DataFrame, busca no DataFrame do clima informado o valor correspondente àquela região naquele ano:

````python
def get_climate_value_of_region(row: pd.Series, climate_dataset: pd.DataFrame):
    region = row['origin_country'].lower()
    target_year = str(row['review_date'].year)

    # Encontrando a linha correspondente à região
    region_info = climate_dataset[climate_dataset['region'] == region]
    if region_info.empty:
        return None

    # Obtendo o valor correspondente à região e ao ano da avaliação
    # Arredonda para 2 casas decimais
    climate_value = round(region_info[target_year].values[0], 2)

    return climate_value
````

A função ***get_flavors_by_hierarchy*** recebe o dicionário encadeado de notas sensoriais e o nível de hierarquia desejado, e retorna a lista de notas daquele nível:

````python
def get_flavors_by_hierarchy(flavors: dict[str, dict[str, list[str]]], hierarchy_level: int) -> list[str]:
    hierarchy_flavors = []

    # Retorna os níveis de hierarquia 0 (as chaves principais)
    if hierarchy_level == 0:
        return list(flavors.keys())

    # Retorna os níveis de hierarquia 1 para cada chave principal
    if hierarchy_level == 1:
        for key, value in flavors.items():
            for sub_key, sub_value in value.items():
                hierarchy_flavors.append(sub_key)
        return hierarchy_flavors

    # Retorna os níveis de hierarquia 3 para cada chave secundária
    if hierarchy_level == 2:
        for key, value in flavors.items():
            for sub_key, sub_value in value.items():
                hierarchy_flavors += sub_value
        return hierarchy_flavors

    return None
````

A função ***get_subflavors*** recebe o dicionário encadeado de notas sensoriais, e retorna as notas sensoriais do próximo subnível da nota informada:

````python
def get_subflavors(flavors: dict[str, dict[str, list[str]]], flavor: str, hierarchy_level: int) -> list[str]:

    if hierarchy_level == 0:
        if flavor in flavors.keys():
            return list(flavors[flavor].keys())

    if hierarchy_level == 1:
        for key, value in flavors.items():
            if flavor in value.keys():
                return value[flavor]

    return None
````

A função ***count_flavor_in_row_desc***, para cada série do DataFrame, busca nas colunas de Informações Extras a quantidade de vezes que é mencionada uma determinada nota sensorial e seus sinônimos:

````python
def count_flavor_in_row_desc(row: pd.Series, flavor: str) -> int:
    synonyms = flavor.split('+')
    flavor_intensity = 0

    # Para cada sinônimo daquela nota sensorial, calcula o valor total de ocorrências delas
    for unique_flavor in synonyms:
        flavor_intensity += str(row['desc_1']).count(unique_flavor)
        flavor_intensity += str(row['desc_2']).count(unique_flavor)
        flavor_intensity += str(row['desc_3']).count(unique_flavor)

    return flavor_intensity
````

A função ***create_flavors_columns*** recebe um DataFrame e o dicionário encadeado, e cria as colunas de cada nota sensorial. Os valores de cada coluna representam a quantidade de aparições da nota sensorial naquela série, acumulando com as aparições das respectivas notas sensoriais do nível inferior àquela.
<br><br>
Por exemplo, a coluna respectiva à nota sensorial 'Berry' terá seus valores de acordo com a quantidade de aparições da palavra Berry somada a quantiade de aparições de 'Blackberry', 'Raspberry', 'Blueberry', 'Strawberry':
```python
flavors = {
    'Fruity': {
        'Berry': ['Blackberry', 'Raspberry', 'Blueberry', 'Strawberry']
    }
}
```

````python
# Função para criar as colunas dinamicamente
def create_flavors_columns(df: pd.DataFrame, flavors: dict[str, dict[str, list[str]]]) -> pd.DataFrame:

    # Para cada nível de hierarquia das notas sensoriais, do maior para o menor
    for hierarchy in reversed(range(FLAVORS_HIERARCHY_LEVELS)):

        # Combina as notas sensoriais, com o título para sua coluna e com o texto o nome da nota sensorial mínuscula
        hierarchy_flavors = get_flavors_by_hierarchy(flavors, hierarchy)
        columns_name = format_flavors_by_hierarchy(hierarchy_flavors, hierarchy)
        lower_flavors = convert_list_itens_to_lower(hierarchy_flavors)
        columns_zip = zip(columns_name, hierarchy_flavors, lower_flavors)

        for column in columns_zip:
            column_name = column[0]
            column_flavor = column[1]
            lower_column_flavor = column[2]

            # Se for o maior nível, cria as colunas contando as aparições de cada flavor
            if hierarchy == (FLAVORS_HIERARCHY_LEVELS - 1):
                df[column_name] = df.apply(count_flavor_in_row_desc, axis=1, args=(lower_column_flavor,))
                continue

            # Se for os níveis menors, que possuem subníveis, cria as colunas acumulando os valores dos subníveis e soma as aparições de cada flavor
            sub_flavors = get_subflavors(flavors, column_flavor, hierarchy)
            sub_columns = format_flavors_by_hierarchy(sub_flavors, hierarchy + 1)
            df[column_name] = df[sub_columns].sum(axis=1)
            df[column_name] += df.apply(count_flavor_in_row_desc, axis=1, args=(lower_column_flavor,))

    return df
````

A função **assign_class_based_on_max_value** recebe uma lista de colunas, referente aos grupos de notas sensoriais do nível sendo analisado e para cada linha, retorna uma tupla com o índice do grupo com maior valor de referências e o próprio valor.

Obs: Esta é a função utilizada para determinar a classe mais representativa do café (*Multi-Class*)

````python
def assign_class_based_on_max_value(row: pd.Series, columns: list[str]) -> tuple[int, int]:
    max_col = max(columns, key=lambda col: row[col])
    max_col_value = row[max_col]
    return int(columns.index(max_col)), int(max_col_value)
````
